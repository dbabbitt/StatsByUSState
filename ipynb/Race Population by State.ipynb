```python

def get_black_population(state, year):
    mask_series = (STATE_RACE_YEAR_DF.index.get_level_values(0) == str(year))
    mask_series = mask_series & (STATE_RACE_YEAR_DF.index.get_level_values(1) == state)
    black_population = STATE_RACE_YEAR_DF.loc[mask_series, 'Black_or_African_American'].squeeze()
    
    return black_population
```

```python

def get_white_population(state, year):
    mask_series = (STATE_RACE_YEAR_DF.index.get_level_values(0) == str(year))
    mask_series = mask_series & (STATE_RACE_YEAR_DF.index.get_level_values(1) == state)
    white_population = STATE_RACE_YEAR_DF.loc[mask_series, 'White'].squeeze()
    
    return white_population
```
