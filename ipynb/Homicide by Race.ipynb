@oudeicrat explains how he came to the values:

"basically I loaded those two CSV datasets into two SQL tables (the crime data and the census data), then I separated each of the census data rows into 4 separate rows for each race (asian+pacific, black, native, white) and then calculated the number of homicides for each of the census rows with distinct year,state,race by summing the victim counts from the crime data table according to the year,state,race. Afterwards I calculated the per100k number  = 100000\*homicides/population for each year,state,race"

"I had to modify some of the data, for example the crime data had 'rhodes island' instead of 'rhode island' and often the number of victims was 0, so I assumed 1 instead. I ignored the unknown race data and the multiple race data"

```python

# https://twitter.com/oudeicrat/status/1331560741862858753?s=20
if s.pickle_exists('RACE_HOMICIDE_RATE_BY_STATE_YEAR_DF'):
    RACE_HOMICIDE_RATE_BY_STATE_YEAR_DF = s.load_object('RACE_HOMICIDE_RATE_BY_STATE_YEAR_DF')
else:
    %run ../py/create_race_homicide_rate_by_state_year_dataframe.py
    RACE_HOMICIDE_RATE_BY_STATE_YEAR_DF = create_race_homicide_rate_by_state_year_dataframe()
    s.store_objects(RACE_HOMICIDE_RATE_BY_STATE_YEAR_DF=RACE_HOMICIDE_RATE_BY_STATE_YEAR_DF)
RACE_HOMICIDE_RATE_BY_STATE_YEAR_DF
```

```python

def get_homicide_rate(state, year, race):
    mask_series = (RACE_HOMICIDE_RATE_BY_STATE_YEAR_DF.index.get_level_values(0) == str(year))
    mask_series = mask_series & (RACE_HOMICIDE_RATE_BY_STATE_YEAR_DF.index.get_level_values(1) == state)
    mask_series = mask_series & (RACE_HOMICIDE_RATE_BY_STATE_YEAR_DF.Perpetrator_Race == race)
    homicide_rate = RACE_HOMICIDE_RATE_BY_STATE_YEAR_DF.loc[mask_series, 'Homicide_Rate_Per_100_000'].squeeze()
    
    return homicide_rate
```

```python

# African–Americans as percentage of total population (1790–2010) by U.S. state
black_pop_df = tables_df_list[27]
black_pop_df = black_pop_df.set_index('State/territory')
black_pop_df.columns = [1790, 1800, 1810, 1820, 1830, 1840, 1850, 1860, 1870, 1880, 1890, 1900, 1910, 1920, 1930, 1940, 1950, 1960, 1970,
                        1980, 1990, 2000, 2010]
black_pop_df.head(5).T.head(5).T
```

```python

blacks_max_year_df = black_pop_df[max_year].reset_index()
blacks_max_year_df.columns = ['State', 'Percent']
blacks_max_year_df['Race'] = 'Black'
column_list = ['State', 'Race']
blacks_max_year_groupby = blacks_max_year_df.groupby(by=column_list)
blacks_max_year_df = blacks_max_year_groupby['Percent'].max().map(f).reset_index()
blacks_max_year_df.sample(5).T.head(5).T
```

```python

whites_max_year_df = white_pop_df['Percent_{}'.format(max_year)].reset_index()
whites_max_year_df.columns = ['State', 'Percent']
whites_max_year_df['Race'] = 'White'
column_list = ['State', 'Race']
whites_max_year_groupby = whites_max_year_df.groupby(by=column_list)
whites_max_year_df = whites_max_year_groupby['Percent'].max().map(f).reset_index()
whites_max_year_df.head(5).T.head(5).T
```

```python

column_list = ['State', 'Race', 'Percent']
percents_df = pd.merge(left=whites_max_year_df, right=blacks_max_year_df, how='outer',
                       on=column_list).sort_values(column_list).reset_index(drop=True)
percents_df.head(5).T.head(5).T
```

```python

column_list = ['State', 'Race']
homicides_max_year_df = pd.merge(left=homicides_max_year_df, right=percents_df, how='outer',
                                 on=column_list).sort_values(column_list).reset_index(drop=True)
homicides_max_year_df.head(5).T.head(5).T
```
