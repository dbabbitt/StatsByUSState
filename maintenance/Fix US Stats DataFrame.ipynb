# Create a Military-expenditures Equivalent Column

```python

military_expenditures_df = s.load_object('military_expenditures_df')
mask_series = (military_expenditures_df.country_name == 'United States')
us_military_expenditures = military_expenditures_df[mask_series].spending_usd_2021.squeeze()
```

```python

us_stats_df = s.load_object('us_stats_df')
gdp_sum = us_stats_df.gdp_millions_usd_2021.sum()
us_stats_df['proportion_of_gdp_2021'] = us_stats_df.gdp_millions_usd_2021.map(lambda x: float(x)/gdp_sum)
us_stats_df['gdp_proportion_of_military_expenditures_2021'] = us_stats_df.proportion_of_gdp_2021.map(lambda x: x*us_military_expenditures)
s.store_objects(us_stats_df=us_stats_df)
columns_list = ['gdp_millions_usd_2021', 'national_precentage_2021', 'proportion_of_gdp_2021', 'gdp_proportion_of_military_expenditures_2021']
us_stats_df[columns_list]
```

```python

# Get the military expenditure equivalents
import numpy as np

eurasia_df = s.load_object('eurasia_df')
country_tuples_list = [(r.country_name, r.military_spending_usd_2021) for i, r in eurasia_df.iterrows()]
us_stats_df = s.load_object('us_stats_df')
state_tuples_list = [(r.state_name, r.gdp_proportion_of_military_expenditures_2021) for i, r in us_stats_df.reset_index().iterrows()]
rows_list = []
for state_tuple in state_tuples_list:
    candidate_tuple = sorted([s for s in country_tuples_list], key=lambda x: abs(x[1] - state_tuple[1]))[0]
    state_name = state_tuple[0]
    country_name = candidate_tuple[0]
    # print(f'{state_name} is close to the military expenditure of {country_name}')
    row_dict = {}
    state_name = state_tuple[0]
    row_dict['state_name'] = state_name
    row_dict['country_name'] = country_name
    rows_list.append(row_dict)
state_to_country_mee_dict = pd.DataFrame(rows_list).set_index('state_name').country_name.to_dict()
us_stats_df['Country_Equivalent_Military_Expenditure'] = us_stats_df.index.map(lambda x: state_to_country_mee_dict.get(x, np.nan))
s.store_objects(us_stats_df=us_stats_df)
columns_list = ['Country_Equivalent_GDP', 'Country_Equivalent_Military_Expenditure']
us_stats_df[columns_list]
```
