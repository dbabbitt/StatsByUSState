```python

from datetime import datetime

expectency_df = s.load_csv(csv_name='united-states-population-2021-04-14', folder_path=None)
expectency_df.date = expectency_df.date.map(lambda x: datetime.strptime(x, "%m/%d/%Y").date())
expectency_df['Year'] = expectency_df.date.map(lambda x: x.year)
expectency_df = expectency_df.set_index('Year')
expectency_df.columns = ['date', 'average_life_expectency', 'annual_percent_change']
expectency_df = expectency_df['average_life_expectency']
```

```python

mask_series = (expectency_df.index >= 1960) & (expectency_df.index <= 2022)
```

```python

expectency_df[mask_series].plot.line();
```

```python

expectency_df[mask_series].diff().plot.line();
```
